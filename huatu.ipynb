```python
from graphviz import Digraph
```

```python
edges = []
```

```python
dot = Digraph()
with open('simplegraph1.txt') as file:
    node_num = int(file.readline())

    for i in range(0,node_num):
        dot.node(str(i))

    data2 = file.readlines()
    for line in data2:
        dot.edge( str(line.split("\n")[0].split(',')[0]), str(line.split("\n")[0].split(',')[1]))
```
